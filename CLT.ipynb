```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Setup random seed for reproducibility
np.random.seed(42)

# 2. Create a heavily skewed population (Exponential distribution)
population_size = 100000
population = np.random.exponential(scale=2, size=population_size)

# 3. CLT Simulation Parameters
num_samples = 1000  # Number of times we draw samples
sample_size = 40  # Size of each sample (n >= 30 rule)

# 4. Collect the mean of each sample
sample_means = []
for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size)
    sample_means.append(np.mean(sample))

# 5. Plot the results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot 1: The Raw Population (Not Normal)
ax1.hist(population, bins=100, color="coral", edgecolor="black", alpha=0.7)
ax1.set_title("1. Original Population\n(Highly Skewed)")
ax1.set_xlabel("Value")
ax1.set_ylabel("Frequency")

# Plot 2: Distribution of Sample Means (Normal!)
ax2.hist(sample_means, bins=30, color="skyblue", edgecolor="black", alpha=0.7)
ax2.set_title(
    f"2. Distribution of {num_samples} Sample Means\n(Bell-shaped / Normal)"
)
ax2.set_xlabel("Sample Mean")
ax2.set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
